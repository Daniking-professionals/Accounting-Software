```python
# List of fowls on the farm with their screening results
# 'Healthy' means no infection, 'Flu' means they tested positive
fowl_screening_results = [
    "Healthy", "Flu", "Healthy", "Healthy", "Flu",
    "Healthy", "Flu", "Flu", "Healthy", "Healthy"
]

# Initialize our counter at 0
infected_count = 0
total_fowls = len(fowl_screening_results)

# Loop through each fowl's result in the list
for status in fowl_screening_results:
    if status == "Flu":
        infected_count += 1  # Add 1 to the count if they have the flu

# Calculate healthy fowls for additional reporting
healthy_count = total_fowls - infected_count

# Display the screening summary
print("=== FARM FLU SCREENING REPORT ===")
print(f"Total Fowls Screened: {total_fowls}")
print(f"Total Infected Fowls (Flu): {infected_count}")
print(f"Total Healthy Fowls: {healthy_count}")
print("---------------------------------")

# Provide a quick action recommendation based on the results
if infected_count > 0:
    print(f"ACTION REQUIRED: Immediately isolate the {infected_count} infected fowl(s) to prevent further spread.")
else:
    print("ALL CLEAR: No flu infections detected in this batch.")
```
