```python
# Create a function that will be used to calculate the bank statement balance

def Balance_as_per_bank_statement(
        adjusted_balance_as_per_cashbook,
        unpresented_cheques,
        errors_of_overstatement_by_bank,
        uncredited_cheques,
        error_reducing_the_business_balance):

    balance = (
        adjusted_balance_as_per_cashbook
        + unpresented_cheques
        + errors_of_overstatement_by_bank
        - uncredited_cheques
        - error_reducing_the_business_balance
    )

    return balance


# Call the function

result = Balance_as_per_bank_statement(
    5115,
    1465,
    -515,
    0,
    0
)

print(result)
```
