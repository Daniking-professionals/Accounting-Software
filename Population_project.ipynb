```python
population_size = 160000000 # initial population in million
growth_rate = 1.05 # 5% annual growth
year = 0
while population_size < 180000000:
    population_size *= growth_rate
    year += 1
    print(f"Population size after {year} years: {population_size:.2f} million")
```
